# Combine Scrape results
**1** I had some goal: to create a table where all films have a country origin.
**2** This meant combining the results of BBC and IMDb scrapes
**3** I created a series of tables, added some info based on grouping and counts (ultimately it was unuseful)
**4** Then I merged the tables. Ultimately I would need properties.name column. At this stage it was cn_of_org

```python
import pandas as pd
```

```python
df1 = pd.read_csv('juhana_BBC02.csv')
```

```python
df1.head()
```

```python
df2 = pd.read_excel('Single_Origin.xlsx')
```

```python
df2
```

```python
df3 = df2.drop(columns=['Column1'])
```

```python
final_df = pd.merge(df1, df3, how ='left', on='movie')
```

```python
final_df
```

```python
final_df.to_csv('movie_master_table.csv')
```

```python
crit_rank1_df = final_df.query("crit_rank == 1")
```

```python
crit_rank1_df.groupby('movie')['crit_rank'].sum()
```

```python
crit_rank1_df.sort_values(by='total_ones')
```

```python
rank1_total_df = crit_rank1_df.groupby('movie')['crit_rank'].sum()
```

```python
new_df = rank1_total_df.to_frame(name='total_ones')
new_df.sort_values(by='total_ones', ascending=False)
```

```python
with_total1s_df = pd.merge(final_df, new_df, how ='left', on='movie')
```

```python
with_total1s_df
```

```python
with_total1s_df['total_ones'].fillna(0, inplace=True)
```

```python
with_total1s_df
```

```python
with_total1s_df.total_ones = with_total1s_df.total_ones.astype(int)
```

```python
with_total1s_df
```

```python
with_total1s_df.to_csv('movie_master_table.csv')
```
